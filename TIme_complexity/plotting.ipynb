```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
import pandas as pd
```

```python
x=[]
y=[]
base_speed=2e+9
```

```python
with open('time_selection.txt', 'r') as file:  
    for line in file:
        parts = line.strip().split()  
        time = float(parts[0].split('=')[1])  
        size = float(parts[1].split('=')[1])  
        
        y.append(time*base_speed)  
        x.append(size)  
```

```python
def test(x,a,b):
    return a*np.square(x)+b
```

```python
param,param_cov=curve_fit(test,x,y)
```

```python
plt.figure(figsize=(8, 6))  
plt.plot(x, y, marker='x', linestyle='-', color='r', label='Time vs Input_Size')
plt.plot(x,test(x,param[0],param[1]),color='g',label='approximate curve',linewidth=3)  
plt.title('Time complexity selection sort')  
plt.xlabel('Input_Size') 
plt.ylabel('CPU Operations')  
plt.legend()  
plt.show()  
```

```python
with open('insertion_time.txt', 'r') as file:  
    for line in file:
        parts = line.strip().split()  
        time = float(parts[0].split('=')[1])  
        size = float(parts[1].split('=')[1])  
        
        y_.append(time*base_speed)  
        x_.append(size)  
```

```python
param_,param_cov_=curve_fit(test,x_,y_)
```
