```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
import pandas as pd
```

```python
x=[]
y=[]
base_speed=2e+9
```

```python
with open('Merge_time.txt', 'r') as file:  
    for line in file:
        parts = line.strip().split()  
        time = float(parts[0].split('=')[1])  
        size = float(parts[1].split('=')[1])  
        
        y.append(time*base_speed)  
        x.append(size)  
```

```python
arr=np.array(x,dtype=float)
```

```python
def test(arr,a,b):
    return a*arr*np.log2(arr)+b
```

```python
param,param_cov=curve_fit(test,arr,y)
```

```python
plt.figure(figsize=(8, 6))  
plt.plot(x, y, marker='x', linestyle='-', color='r', label='Time vs Input_Size')
plt.plot(arr,test(arr,*param),color='g',label='approximate curve',linewidth=3)  
plt.title('Time complexity Merge sort')  
plt.xlabel('Input_Size') 
plt.ylabel('CPU Operations')  
plt.legend()  
plt.show()  
```
